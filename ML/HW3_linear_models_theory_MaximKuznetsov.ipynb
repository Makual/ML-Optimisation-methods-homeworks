(*3 points* )

To prepare for a test, FCS students eat candies and solve problems. The number of solved problems $y$ depends on the number of eaten candies $x$. If a student has not eaten any candy, they do not want to solve problems. Therefore, for describing the dependency of the number of solved tasks on the number of eaten candies, a linear model with one feature and no constant is used: $y_i = w \cdot x_i.$ Find an analytic expression of estimates of the parameter $w$ by minimizing the following loss function:

$$Q(w) = \frac{1}{\ell} \sum_{i=1}^{\ell} (y_i - w x_i)^2.$$

Q(w) = ((y1 - wx1)^2 + (y2 - wx2)^2 + (y3 - wx3)^2 + ...) / l

Q'(w) = 1/l * (2*(y1 - wx1)*x1 + 2*(y2 - wx2)*x2 + 2*(y3 - wx3)*x3 + ...) = 0

2y1*x1 - 2w*x1^2 + 2y2*x2 - 2w*x2^2 + 2y3*x3 - 2w*x3^2 = 0

w*(Z(x_l^2, 1, l, 1)) = Z(x_l*y_l, 1, l, 1)

w = Z(x_l*y_l, 1, l, 1) / Z(x_l^2, 1, l, 1)

# Problem 2

James measured the weight of three Pokemons, $y_1=6$, $y_2=6$, $y_3=10$ (in kg). Now James wants to predict weight for the next Pokemon using a constant model  $y_i = w$. For estimating the parameter $w$ he uses the following target function:

$$\frac{1}{\ell}\sum_{i=1}^{\ell} (y_i - w)^2 + \lambda w^2.$$

1. Find an optimal $w$ for an arbitrary $\lambda$ (*2 points* )
1. Find an optimal $\lambda$ using LOOCV (leave one out cross-validation) (*2 points* )
> LOOCV algorithm utilizes the following logic: On the first step, we evaluate the model on the entire sample without the first observation, then test it on the first observation. On the second step, we evaluate the model on the entire sample without the second observation, and now test it on the second observation. The algorithm is performed $\ell$ times. In order to find $\lambda_{CV}$ we minimize the average error made on the test samples.

1. Find an optimal value of $w$ for $\lambda_{CV}$, found with the algorithm above (*2 points* )
1. Derive an analytical expression for $\lambda_{CV}$ for an arbitrary number of observations (*2 points* )
